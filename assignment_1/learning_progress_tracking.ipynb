```python
week_of_month = ['Week 1','Week 2','Week 3','Week 4'] 
task_list = ['Task 1','Task 2','Task 3','Task 4']

def get_input():
    monthly_progress = []
    task_name = []
    for i in task_list:
        task_name.append(input('Enter ' + i + "'s Name:"))   
    for n in task_name:
        w1_hrs_spent = input('Enter Week 1 ' + n + "'s Hours Spent:")
        w2_hrs_spent = input('Enter Week 2 ' + n + "'s Hours Spent:")
        w3_hrs_spent = input('Enter Week 3 ' + n + "'s Hours Spent:")
        w4_hrs_spent = input('Enter Week 4 ' + n + "'s Hours Spent:")
        monthly_progress.append({'Task Name':n,'Week 1 Hours Spent':w1_hrs_spent,'Week 2 Hours Spent':w2_hrs_spent,'Week 3 Hours Spent':w3_hrs_spent,'Week 4 Hours Spent':w4_hrs_spent})
    return monthly_progress,task_name

def save_csv(ds):
    import pandas as pd
    df = pd.DataFrame(ds)
    df.to_csv('learning_progress.csv', mode='w', index=False)

def read_csv(csv_file):
    import pandas as pd
    df = pd.read_csv(csv_file)
    return df

data, task_name = get_input()
save_csv(data)
dataframe = read_csv('learning_progress.csv')
```

```python
def visualize(df):
    from matplotlib import pyplot as plt
    progress_list = df.values.tolist()
    w1_hrs_spent = []
    w2_hrs_spent = []
    w3_hrs_spent = []
    w4_hrs_spent = []
    for i in progress_list:
        w1_hrs_spent.append(i[1])
        w2_hrs_spent.append(i[2])
        w3_hrs_spent.append(i[3])
        w4_hrs_spent.append(i[4])
        avr_hrs_spent = [sum(w1_hrs_spent)/4,sum(w2_hrs_spent)/4,sum(w3_hrs_spent)/4,sum(w4_hrs_spent)/4] 
    for i in progress_list:
        t1_hrs_spent = progress_list[0][1:]
        t2_hrs_spent = progress_list[1][1:]
        t3_hrs_spent = progress_list[2][1:]
        t4_hrs_spent = progress_list[3][1:]
    plt.plot(week_of_month, t1_hrs_spent, color='red',marker='s') 
    plt.plot(week_of_month, t2_hrs_spent, color='orange',marker='s') 
    plt.plot(week_of_month, t3_hrs_spent, color='yellow',marker='s') 
    plt.plot(week_of_month, t4_hrs_spent, color='green',marker='s') 
    plt.plot(week_of_month, avr_hrs_spent, color='blue',linestyle='--') 
    plt.legend([task_list[0] + ':' + task_name[0],task_list[1] + ':' + task_name[1],task_list[2] + ':' + task_name[2],task_list[3] + ':' + task_name[3],'Weekly Average'],bbox_to_anchor=(1.05, 1.0), loc='upper left')
    plt.xlabel("Week of Month")
    plt.ylabel("Hours Spent")
    plt.title('Learning Progress Tracking')
    plt.show()

visualize(dataframe)
```
