```python
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import re
from bs4 import BeautifulSoup
import requests
import lxml
import time

project_path = os.path.abspath(os.path.dirname('__file__'))
print(project_path)
driver_path = os.path.join(project_path, "chromedriver")
print(driver_path)
driver = webdriver.Chrome(driver_path)
driver.get('https://search.damai.cn/search.htm?spm=a2oeg.search_category.top.dcategory.662e4d15EHWzp3&order=1')
driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[1]/div[1]/div[2]/div/div/div/span[3]').click()
time.sleep(2)
driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[1]/div[1]/div[3]/div/div/div/span[4]').click()
time.sleep(2)
driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[1]/div[2]/div[1]/span[3]').click()
time.sleep(5)
                             
header = ['title','url','date','location','price','artists']
data = []
count = 0

while count<3:
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')
    for item in soup.find_all("div", attrs={"class":"items"}):
        list = []
        title = item.find("div",attrs={"class":"items__txt__title"}).get_text().replace('\n','').replace('\r','').strip()
        url = "https:" + item.a.get('href').replace('\n','').replace('\r','').strip()
        price_i= item.find("div", attrs={"class":"items__txt__price"}).span
        price = price_i.get_text().replace('\n','').replace('\r','').strip() + "元"
        for div in item.find_all("div", attrs={"class":"items__txt__time"}):
            list.append(div.get_text().replace('\n','').replace('\r','').strip())
        if len(list)>2:
            artists = list[0]
            location = list[1]
            date = list[2]
        else:
            artists = ''
            location = list[0]
            date = list[1]
        data.append((title, url, date, location, price, artists))
    driver.find_element_by_class_name('btn-next').click()
    time.sleep(5)
    count = count+1


data
```

```python
import pandas as pd
df = pd.DataFrame(data,
  columns = header
)
df
```

```python
df.to_csv('Damai.csv', sep='\t', encoding='utf-8')
```
