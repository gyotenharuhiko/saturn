```python
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import re
from bs4 import BeautifulSoup
import requests
import lxml
import time

project_path = os.path.abspath(os.path.dirname('__file__'))
print(project_path)
driver_path = os.path.join(project_path, "chromedriver")
print(driver_path)
driver = webdriver.Chrome(driver_path)
driver.get('https://www.showstart.com/event/list?cityCode=10')
elem = driver.find_element_by_tag_name("input")
elem.send_keys('音乐节')
elem.send_keys(Keys.ENTER)
time.sleep(10)
header = ['title','url','date','location','price','artists','images']
data = []
count = 0

while count<3:
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')
    for item in soup.find_all("a", attrs={"class":"show-item item"}):
        list = []
        title = item.find("div",attrs={"class":"title"}).get_text().replace('\n','').replace('\r','').strip()
        url = "https://www.showstart.com" + item.get('href').replace('\n','').replace('\r','').strip()
        artists = item.find("div",attrs={"class":"artist"}).get_text().replace('\n','').replace('\r','').strip()
        date = item.find("div",attrs={"class":"time"}).get_text().replace('\n','').replace('\r','').strip()
        location = item.find("div",attrs={"class":"addr"}).get_text().replace('\n','').replace('\r','').strip()
        images = item.find("img",attrs={"class":"el-image__inner"}).get('src')
        price = item.find("div",attrs={"class":"price"}).span.get_text().replace('\n','').replace('\r','').strip()
        data.append((title, url, date, location, price, artists, images))
    driver.find_element_by_class_name('btn-next').click()
    time.sleep(5)
    count = count + 1
data
```

```python
import pandas as pd
df = pd.DataFrame(data,
  columns = header
)
df
```

```python
df.to_csv('ShowStart.csv', sep='\t', encoding='utf-8')
```

```python
import os
import requests
images = df['images'].values.tolist()
dir_name = 'music_festival'
if not os.path.exists(dir_name):     
        os.mkdir(dir_name)
for i in range(len(images)):
    r = requests.request('get',images[i])
    with open(dir_name + str('image') + str(i)+'.png','wb') as f:
        f.write(r.content)
    f.close()
```
